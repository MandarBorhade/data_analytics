```python
import pandas as pd
```

```python
data_import = pd.read_csv("/content/data_cleaning_challenge.csv")
```

```python
data_import.head()
```

```python
data_import.isna().sum()
```

```python
data_import.info()
```

```python
#dropping colums Unnamed: 9 & Unnamed: 10
drop_extra = data_import.drop(columns = ['Unnamed: 9' , 'Unnamed: 10'])
```

```python
drop_nans = drop_extra[drop_extra['Row Type'].notna()]
drop_nans.isna().sum()
```

```python
drop_nans.head()
```

```python
column_vals = []
counter = 0
for i in drop_nans['Row Type']:
  if 'first name' in i:
    counter = counter + 1
  column_vals.append(counter)
```

```python
iter_cols = drop_nans
iter_cols['Iteration'] = column_vals
iter_cols.head()
```

```python
iter_cols
```

```python
drop_extra_column_name_rows = iter_cols[iter_cols['Row Type'] != 'Row Type']
drop_extra_column_name_rows
```

```python
name_dataframe = drop_extra_column_name_rows[drop_extra_column_name_rows['Row Type'].str.contains('first name')]
name_dataframe
```

```python
name_dataframe = name_dataframe.drop(columns = ['Speed1' , 'Speed2' , 'Electricity' , 'Effort' , 'Weight' , 'Torque'])
name_dataframe
```

```python
name_dataframe.rename(columns = {'Row Type':'First Name' , 'Iter Number':'Last Name' , 'Power1':'Date' } , inplace=True)
name_dataframe['First Name'] = name_dataframe['First Name'].str[12:]
name_dataframe['Last Name'] = name_dataframe['Last Name'].str[11:]
name_dataframe['Date'] = name_dataframe['Date'].str[6:]
name_dataframe
```

```python
no_name_dataframe = drop_extra_column_name_rows[~drop_extra_column_name_rows['Row Type'].str.contains('first name')]
no_name_dataframe
```

```python
pd.merge(left=name_dataframe , right=no_name_dataframe , how='inner' , on='Iteration')
```
